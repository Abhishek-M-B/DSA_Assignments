Q1. Write a Python program which accepts the radius of a circle from the user
and compute the area.

```python
radius=int(input("Enter the radius: "))
pi=3.14
Area=pi*radius**2
print(f"Area of a circle is: {Area}")
```

Q3. Write a Python program to get the largest number from a list.
Sample input: 12, 3, 47, 10

```python
list1 = input('enter numbers: ').split(",")

for i in range(len(list1)):
  list1[i] = int(list1[i])

print(f"Largest number is : {max(list1)}")
```

Q4.  Given a range of first 10 numbers, write a Python program to iterate from
start number to the end number and print the sum of the current number
and previous number.  


```python
prev_num = 0

for curr_num in range (1,11):
  sum = curr_num +prev_num
  print(f'Current Number {curr_num} Previous Number {prev_num} Sum: {sum}')
  prev_num = curr_num
```

Q5. Write a Python program to print only those numbers which are divisible
of 5.

```python
nums = list(map(int, input("Enter numbers separated by space: ").split()))

for n in nums:
    if n % 5 == 0:
        print(n,end=" ")
```

Q6. Write a Python program to check whether a number is prime or not.

```python
num=int(input("Enter the number: "))
if num <= 1:
    print("Number less than 1 is not a prime number")
else:
  for i in range(2,num):
     if num%i==0:
       print(f"{num} not a prime number")
       break
  else:
     print(f"{num} is prime number")
```

Q9. Write a Python program to find the maximum of three numbers  
Sample input: 34,12,7  

```python
numbers = [34, 12, 7]

maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

print("Maximum number:", maximum)
```

Q10.  Write a Python program to construct the following pattern, using a
nested for loop.

```python
num = int(input("Enter the number: "))

for i in range(1, num + 1):
    for k in range(i):
        print("*",end="")
    print()

for j in range(num - 1, 0, -1):
    for k in range(j):
        print("*",end="")
    print()
```
