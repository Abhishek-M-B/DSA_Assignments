### 4) Check and remove duplicates

```python
# Check for duplicates
print("Number of Duplicate Rows:", df.duplicated().sum())

# Remove duplicate rows and update the DataFrame in-place
df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", df.shape)
```

### 5) Rename columns ("Title" > "Movie_Title")

```python
df=df.rename(columns={"Title":"Movie_Title"})
df
```

### 6) Get unique genres

```python
df["Genre"].unique()
```

### 7) Extract number of genres per movie

```python
df["no_of_genres"]=df["Genre"].str.split(",").str.len()
df
```

### 8)  Handle missing values in 'Revenue (Millions)' and 'Metascore'

```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())
df[['Revenue (Millions)', 'Metascore']].isna().sum()
```

### 9) Top 5 movies by Rating

```python
top_5_movies = df.nlargest(5, 'Rating')
print(top_5_movies[['Title', 'Rating']])
```

### 10) Movies with highest/lowest revenue

```python
highest_revenue_movie = df.nlargest(1, 'Revenue (Millions)')
lowest_revenue_movie = df.nsmallest(1, 'Revenue (Millions)')

print(highest_revenue_movie[['Title', 'Revenue (Millions)']])
print(lowest_revenue_movie[['Title', 'Revenue (Millions)']])
```

### 11)  Avg. rating by Director

```python
avg_rating_by_director = df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
avg_rating_by_director
```

### 12) Movies per year

```python
movies_per_year = df['Year'].value_counts().sort_index(ascending=False)
movies_per_year
```

### 13) Movies with rating > 8 & revenue > 100M

```python
Rating_above_8=df[(df["Rating"]>8) & (df["Revenue (Millions)"]>100)][["Title","Rating","Revenue (Millions)"]].sort_values(by="Rating",ascending=False)
Rating_above_8
```

### 14) Group by Director: avg Rating, Revenue

```python
director_summary = df.groupby('Director')[['Rating', 'Revenue (Millions)']].mean()
director_summary
```

### 15) Group by Year: count, avg Rating

```python
year_status = df.groupby("Year").agg(
    movie_count=("Title", "count"),
    avg_rating=("Rating", "mean")
)

year_status
```

### 16) Create a new column Profitability:
 a) Revenue divided by Duration
 b) Find top 10 most profitable movies

```python
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
top_10=df.nlargest(10,"Profitability")
top_10[["Title","Revenue (Millions)","Runtime (Minutes)","Profitability"]]
```

### 17) Create a column that categorizes movies by rating:
 a) Rating < 6: Poor
 b) 6 â‰¤ Rating < 7.5: Average
 c) Rating â‰¥ 7.5: Excellent

```python
df["Rating_Summary"]=df["Rating"].apply(lambda x: 'Poor' if x < 6 else ('Average' if x < 7.5 else 'Excellent'))
df[['Title', 'Rating', 'Rating_Summary']]
```

### 18) Create a custom function to score movies based on:  
a) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
b) Apply a function to dataset to assign the score

```python

df['High_Score'] = df.apply(lambda row: row['Rating'] * 10 + row['Revenue (Millions)'], axis=1)
df[['Title', 'Rating', 'Revenue (Millions)', 'High_Score']]
```
