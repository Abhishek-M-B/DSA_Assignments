```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df
```

```python
df.shape
```

```python
df.nunique()
```

```python
df['species'].unique()
```

```python
#pairplot used to see all together
sns.pairplot(df,hue='species')
```

```python
new_df=df.select_dtypes(include=['number'])
corr=new_df.corr()
corr
```

```python
sns.heatmap(corr,annot=False,cmap="Reds")
```

```python
#A specialized heatmap that performs hierarchical clustering on the data to rearrange the rows and columns. This helps in visually grouping observations or features that are similar to each other.
sns.clustermap(new_df)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=df)
```

```python
#The most basic way to visualize the relationship between two or more numerical variables. You can easily add a third variable using the 'hue' or 'size' parameters.
sns.scatterplot(x='sepal_length', y='petal_length', hue="species",data=df)
```

```python
# KDE can invent patterns if data is small
sns.histplot(df['sepal_length'], kde=True)
```

```python
# Used to estimate and visualize the probability density function (PDF) of numerical data.It shows where data values are concentrated, using a smooth curve instead of bins.
sns.kdeplot(df)
```

```python
sns.boxplot(x='species', y='sepal_length', data=df)
```

```python
# distribution plot factory used to visualize the distribution of a numerical variable.
sns.displot(df)
```

```python
#Combines a box plot with a KDE plot (density). It shows the full distribution shape like a KDE, but also includes the median and quartiles like a box plot. Great for comparing distributions across categories
sns.violinplot(x='species', y='sepal_length', data=df)
```

```python
#Similar to a histogram, but specifically for categorical variables. It shows the count of observations in each category (i.e., it plots the frequency of categories).
sns.countplot(x='species', data=df)
```

```python
#Draws a scatter plot and fits a linear regression model (straight line) through the data, often with a shaded confidence interval around the line. Great for checking the linearity of a relationship.
sns.regplot(x='sepal_length', y='petal_length', data=df)
```

```python
# Used to visualize the relationship between two numeric variables and fit a linear regression model on top of the scatter plo
sns.lmplot(x='sepal_length', y='petal_length', hue='species', data=df)
```

```python
# Used to visualize how a numerical variable behaves across categorical groups, with support for comparing many such plots at once.
sns.catplot(x='species', y='sepal_length', data=df, kind='box')
```

```python
# Used to analyze the relationship between two numerical variables and their individual distributions at the same time.
sns.jointplot(df)
```

```python
#Empirical Cumulative Distribution Function | Plots the proportion of data that is less than or equal to a given value. Useful for comparing two distributions and seeing the cumulative frequency.
sns.ecdfplot(df)
```

```python
#Best used for visualizing how one variable changes over another, typically time or an ordered category. Excellent for time-series data.
sns.lineplot(df)
```

```python
# Similar to a strip plot, but adjusts the points so they don't overlap, creating a "swarm" effect. A great alternative to a strip plot for showing individual observations.
sns.swarmplot(df)
```

```python
#Plots the individual data points for a numerical variable within each category. Useful for small datasets or when combined with a box plot to show the underlying data.
sns.stripplot(df)
```
