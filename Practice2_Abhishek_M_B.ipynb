Q8.If you invest $500 with a return rate of 4% per year for 3 years, calculate and
print the total amount you will have using simple interest

```python
P=500  # P = Principal (initial amount invested)
R=4  # R = Rate of interest (per year, in %)
T=3  # T = Time (in years)

SI=(P*R*T)/100

total_amount=P+SI

print(f"Simple interest is: {SI}")
print(f"Total amount received with SI: {total_amount}")

```
