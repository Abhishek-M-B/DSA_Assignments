# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_max(numbers):
  maximum=numbers[0] # sett first val of list as maximum
  minimum=numbers[0] # sett first val of list as minimum

  for num in numbers:
    if num >= maximum:
      maximum=num
    elif num <= minimum:
      minimum=num

  print(f"Maximum number: {maximum}")
  print(f"Minimum number: {minimum}")
  print(f"Range : {maximum-minimum}")


numbers=[2,3,4,5]
find_max(numbers)
```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean_val(numbers):
  sum=0
  for number in numbers:
    sum +=number

  total_count=len(numbers)
  mean = sum/total_count
  return mean

num=[1,2,3,4,5,6]
print(mean_val(num))
```

# Q6. Write a function that returns the lower bound of a list of numbers

```python
def median_val(number):
  number.sort()

  avg=len(number)//2

  if len(number)%2==0:
    mid=(number[avg] + number[avg-1])/2
    q1=number[mid-2]
    q3=number[mid+2]
    IQR=q3-q1
    lower_bound=q1-1.5*IQR
    print(f"Lower bound: {lower_bound}")


  else:
    mid=numbers[avg]
    q1=number[mid-2]
    q3=number[mid+2]
    IQR=q3-q1
    lower_bound=q1-1.5*IQR
    print(f"Lower bound: {lower_bound}")


numbers=[1,2,3,4,5,6,7]
median_val(numbers)



```

##Q8. Write a function to return standard deviation of a list of numbers

```python
def SD(num):
    mean = sum(num) / len(num)
    sqr = 0

    for i in num:
        sqr += (i - mean) ** 2

    variance = sqr / len(num)
    sd = variance ** 0.5
    return sd

num = [1, 2, 3, 4, 5]
print(SD(num))
```
